```python
import geopandas as gpd
from shapely.geometry import Point
```

```python
gdf_localidades=gpd.read_file("/content/drive/MyDrive/POO/BarriosComunaSoacha.geojson")
gdf_localidades.to_crs(epsg=9377,inplace=True)
gdf_localidades.plot()
```

```python
gdf_localidades
```

```python
mapeo = {1: 'Comuna 1', 2: 'Comuna 2', 3: 'Comuna 3', 4: 'Comuna 4', 5: 'Comuna 5', 6: 'Comuna 6'}
gdf_localidades['COMUNA'] = gdf_localidades['COMUNA'].replace(mapeo)
gdf_localidades
```

```python
gdf_localidades.plot()
```

```python
mapa = gdf_localidades.plot(
    column="COMUNA",
    figsize=(10, 6),
    edgecolor="black",
    cmap="Set1",
    legend=True,
    legend_kwds={
        'loc':'upper left'}
)
```
